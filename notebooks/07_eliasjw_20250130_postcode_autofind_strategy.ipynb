```python
for d in test_datasets:
    print(f"dataset: {d}")
    df = catalog.load(d)
    results = calculate_all_match_success(df)
    display(results)
```

```python
for d in test_datasets:
    print(f"dataset: {d}")
    df = catalog.load(d)
    results = calculate_all_match_success(df)
    display(results)
    df_postcode, best_match_success, converted = auto_convert_postcode_column(df, results)
    print(best_match_success)
    if converted:
        display(df_postcode["FORMATTED_POSTCODE"])
    else:
        print("Conversion failed")
```
