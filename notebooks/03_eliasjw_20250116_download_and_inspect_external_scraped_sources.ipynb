## Check for completeness with other two data sources

open_data_postal_code

```python
def compare_postal_codes(
    df1, df2, df1_alias, df2_alias, postal_code_col="postal_code_merge_key"
):
    # Get unique postal codes from each DataFrame
    df1_postal_codes = set(df1[postal_code_col].unique())
    df2_postal_codes = set(df2[postal_code_col].unique())

    # Count the number of unique postal codes in each DataFrame
    df1_postal_count = len(df1_postal_codes)
    df2_postal_count = len(df2_postal_codes)

    # Calculate the total number of unique postal codes
    total_postal_count = len(df1_postal_codes.union(df2_postal_codes))

    # Calculate the percentage of postal codes in each DataFrame
    df1_percentage = (df1_postal_count / total_postal_count) * 100
    df2_percentage = (df2_postal_count / total_postal_count) * 100

    # Print the results
    print(f"{df1_alias} has {df1_postal_count} postal codes ({df1_percentage:.2f}%)")
    print(f"{df2_alias} has {df2_postal_count} postal codes ({df2_percentage:.2f}%)")
```

```python
df1 = onemap_merge
df1_alias = "onemap"
df2 = postal_merge
df2_alas = "Opendata"
compare_postal_codes(df1, df2, df1_alias, df2_alas)
```

```python
df1 = onemap_merge
df1_alias = "onemap"
df2 = postcode_address_merge
df2_alas = "postcode_address"
compare_postal_codes(df1, df2, df1_alias, df2_alas)
```

```python
df1 = postal_merge
df1_alias = "Opendata"
df2 = postcode_address_merge
df2_alas = "postcode_address"
compare_postal_codes(df1, df2, df1_alias, df2_alas)
```

## Testing accuracy of postal code matching compared to onemap

```python
postal_codes_comp.columns
```

postal_codes_comp.columns

```python
import geopandas as gpd

geo1 = gpd.GeoDataFrame(
    postal_codes_comp,
    geometry=gpd.points_from_xy(
        postal_codes_comp["LONGITUDE"], postal_codes_comp["LATITUDE"]
    ),
    crs="EPSG:4326",
).to_crs("EPSG:3857")
geo2 = gpd.GeoDataFrame(
    postal_codes_comp,
    geometry=gpd.points_from_xy(
        postal_codes_comp["longitude"], postal_codes_comp["latitude"]
    ),
    crs="EPSG:4326",
).to_crs("EPSG:3857")
```

```python
postal_codes_comp = postal_codes_comp.assign(geo_dist=geo1.distance(geo2))
postal_codes_comp.sort_values("geo_dist", ascending=False)[
    [
        "place_name",
        "ADDRESS",
        "geo_dist",
        "postal_code_merge_key",
        "latitude",
        "longitude",
        "LATITUDE",
        "LONGITUDE",
    ]
].head(25)
```

Based on the above, we can see that the postal codes from the OneMap API and the OpenData postal codes are not a perfect match. The postal codes from the OneMap API are more accurate as they are based on the actual latitude and longitude coordinates of the postal codes. The OpenData postal codes are based on the postal code itself, which may not always be accurate. We can use the OneMap API to get the latitude and longitude coordinates of the postal codes and use that information to update the OpenData postal codes. It could be that there is geocoding for the address plus postal codes with the first data source. Whereas onemap should be more accurate.

## Final pipeline

Use OneMap, for NaNs, combine Opendata with Postcode base.

```python
onemap  # assign SOURCE = "OneMap"
```

```python
import pandas as pd

def get_postal_codes(postal_codes):
    return postal_codes.astype("string").str.zfill(6)


def format_onemap(df):
    return df.assign(
        POSTAL=get_postal_codes(df["POSTAL"]),
        SOURCE="OneMap",
        URL="https://www.onemap.gov.sg/apidocs/search",
        URL_2="https://github.com/xuancong84/singapore-address-heatmap/blob/master/database.csv.gz",
    ).drop(columns=["X", "Y"])


def format_opendata(df):
    df = df.loc[df["country_code"] == "SG"]
    df = df.assign(postal_code=get_postal_codes(df["postal_code"]))
    df = df.assign(
        ADDRESS=(df["place_name"] + " singapore " + df["postal_code"])
        .str.replace(",", "")
        .str.upper(),
        ROAD_NAME=df["place_name"].str.replace(",", "").str.upper(),
    )
    return df.rename(
        columns={
            "postal_code": "POSTAL",
            "latitude": "LATITUDE",
            "longitude": "LONGITUDE",
        },
        inplace=False,
    )[["ADDRESS", "ROAD_NAME", "LATITUDE", "LONGITUDE", "POSTAL"]].assign(
        SOURCE="opendatasoft",
        URL="https://public.opendatasoft.com/explore/dataset/geonames-postal-code/table/?refine.country_code=SG",
    )


def format_postcodebase(df):
    df = df.assign(postcode=get_postal_codes(df["postcode"]))
    df = df.assign(
        ADDRESS=(
            df["address"].str.replace(r" is located in Singapore.*$", "", regex=True)
            + " singapore "
            + df["postcode"]
        )
        .str.replace(",", "")
        .str.upper()
    )
    return df.rename(
        columns={
            "postcode": "POSTAL",
        },
        inplace=False,
    )[["ADDRESS", "POSTAL"]].assign(
        SOURCE="Postcodebase",
        URL="https://sgp.postcodebase.com/all",
    )


def enrich_open_postcode(open_postcode, postcode_base):
    return open_postcode.drop(columns=["ADDRESS"]).merge(
        postcode_base,
        left_on="POSTAL",
        right_on="POSTAL",
        how="left",
        validate="m:1",
        suffixes=("", "_2"),
    )


def extend_onemap(onemap_postal_codes, open_postcode_enriched):
    additional_postal_codes = open_postcode_enriched.loc[~open_postcode_enriched["POSTAL"].isin(onemap_postal_codes["POSTAL"].values)].reset_index(drop=True)
    onemap_postal_codes = pd.concat([onemap_postal_codes, additional_postal_codes], ignore_index=True, axis=0)
    return onemap_postal_codes


def postcode_full_type_conversion(df):
    rename_columns = {
            "POSTAL": "string",
            "ADDRESS": "string",
            "BLK_NO": "string",
            "BUILDING": "string",
            "ROAD_NAME": "string",
            "LATITUDE": "float",
            "LONGITUDE": "float",
            "SOURCE": "string",
            "URL": "string",
            "SOURCE_2": "string",
            "URL_2": "string",
        }
    return df.astype(
        rename_columns
    )[list(rename_columns.keys())]

onemap = catalog.load("one_map_scrape")
open_postcode = catalog.load("open_data_postal_code")
postcode_base = catalog.load("sg_postcode_based_via_getdata")

onemap_postal_codes = format_onemap(onemap)
open_postcode = format_opendata(open_postcode)
postcode_base = format_postcodebase(postcode_base)

open_postcode_enriched = enrich_open_postcode(open_postcode, postcode_base)
onemap_postal_codes_extended = extend_onemap(onemap_postal_codes, open_postcode_enriched).pipe(postcode_full_type_conversion)
onemap_postal_codes_extended
```
